```python
import pandas as pd
from google.cloud import bigquery

# Configuração do Projeto
project_id = 'involuted-disk-479317-a8' # <--- COLOQUE SEU ID AQUI
client = bigquery.Client(project=project_id)

# A Query SQL (Exatamente a que funcionou no BigQuery)
sql = """
SELECT
    date,
    total_confirmed_cases
FROM
    `bigquery-public-data.covid19_italy.data_by_region`
WHERE
    region_name = 'Lombardia'
ORDER BY
    date ASC
"""

# Executa e salva no DataFrame 'df'
df = client.query(sql).to_dataframe()

# Mostra as primeiras 5 linhas para confirmar (Igual ao video!)
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Configura o tamanho do gráfico
plt.figure(figsize=(12, 6))

# Cria o gráfico de linhas (Eixo X = Data, Eixo Y = Casos)
plt.plot(df['date'], df['total_confirmed_cases'], color='red', linewidth=2)

# Adiciona títulos e etiquetas
plt.title('Evolução dos Casos de COVID-19 na Lombardia (Itália)', fontsize=16)
plt.xlabel('Data', fontsize=12)
plt.ylabel('Total de Casos Confirmados', fontsize=12)
plt.grid(True, alpha=0.3) # Adiciona uma grade suave

# Mostra o gráfico
plt.show()
```
